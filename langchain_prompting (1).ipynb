```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
# 1. Load your text file
loader = TextLoader("/content/sample.txt")  # Make sure this file exists in your working directory
documents = loader.load()
```

```python
# 2. Split into chunks
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,        # Number of characters per chunk
    chunk_overlap=50       # Overlap to preserve context
)
chunks = splitter.split_documents(documents)
```

```python

# 3. Print a few chunks
for i, chunk in enumerate(chunks[:5]):
    print(f"\n--- Chunk {i+1} ---\n")
    print(chunk.page_content)
```

```python
from transformers import pipeline
from huggingface_hub import login

# Authenticate using your Hugging Face token
login(token="my_token_id")

# Load the model using pipeline (FLAN-T5 is a text2text model)
generator = pipeline("text2text-generation", model="google/flan-t5-base")

# Define context and question
context = """
LangChain is a framework for developing applications powered by large language models (LLMs).
It enables connecting LLMs to external data, APIs, and chaining multiple calls.
"""
question = "What is LangChain used for?"

# Create a prompt in natural instruction format
prompt = f"""Use the following context to answer the question.
Context: {context}
Question: {question}
Answer:"""

# Generate response
response = generator(prompt, max_new_tokens=100)

# Print result
print(response[0]['generated_text'])
```
